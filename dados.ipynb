```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Carregando o Dataset Titanic
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.tail()
```

```python
# Histograma da idade dos passageiros
plt.hist(dados["Age"].dropna(), bins=30)
plt.title("Faixa Etária")
plt.xlabel("Idade")
plt.ylabel("Frequência")
```

```python
dados["Survived"].value_counts() # 0 - N sobreviveu | 1 - Sobreviveu
```

```python
# Boxplot da idade por sobrevivência
sns.boxplot(y="Age", x="Survived", data=dados, hue="Survived")
plt.title("Idade por Sobrevivência")
plt.xlabel("Sobrevivência (0 = Não | 1 = Sim)")
plt.ylabel("Idade")
```

```python
# Histograma da idade
fig = px.histogram(dados, x="Age", title="Distribuição da Idade")
fig.show()
```
